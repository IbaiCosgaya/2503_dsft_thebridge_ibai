# Apple Stock

### Introduction:

We are going to use Apple's stock price.


### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/09_Time_Series/Apple_Stock/appl_1980_2014.csv)

### Step 3. Assign it to a variable apple and apply a head()

```python
url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/09_Time_Series/Apple_Stock/appl_1980_2014.csv"
apple = pd.read_csv(url)
apple.head()
```

### Step 4.  Check out the type of the columns

```python
apple.dtypes
```

### Step 5. Transform the Date column as a datetime type and check the info()

```python
from datetime import datetime

apple['Date'] = pd.to_datetime(apple['Date'])
apple
```

### Step 6.  Set the date as the index

```python
apple.set_index(['Date'], inplace=True)
apple
```

### Step 7.  Is there any duplicate dates?

```python
duplicados = apple.index.duplicated()
num_duplicados = duplicados.sum()
print(f"Número de índices duplicados: {num_duplicados}")
```

### Step 8.  Ops...it seems the index is from the most recent date. Make the first entry the oldest date.

```python
apple.sort_index()
```

### Step 9. Get the last business day of each month

```python
last_business_days = pd.date_range(start=apple.index.min(), end=apple.index.max(), freq="BM")

mask = apple[apple.index.isin(last_business_days)]

mask
```

### Step 10.  What is the difference in days between the first day and the oldest

```python
oldest = apple.index.min() 
first = apple.index[0]

print((first - oldest).days)
```

### Step 11.  How many months in the data we have?

```python
unique_months = apple.index.to_period('M').nunique()
unique_months
```

### Step 12. Plot the 'Adj Close' value. Set the size of the figure to 13.5 x 9 inches

```python
plt.figure(figsize=(13.5, 9)) 
plt.plot(apple.index, apple['Adj Close'], label='Adj Close', color='b')

plt.xlabel("Date")
plt.ylabel("Adjusted Close Price")
plt.title("Adjusted Close Price Over Time")
plt.legend()
plt.grid(True)

plt.show()
```

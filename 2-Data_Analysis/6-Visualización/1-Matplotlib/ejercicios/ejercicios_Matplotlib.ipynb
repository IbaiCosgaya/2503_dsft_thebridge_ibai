2. Sabemos que podemos pintar gráficas de dos formas: la figura incluye los ejes o teniendo figura y ejes por separado.

Usando solo una figura, usa numpy para los valores del eje X entre 0 y 5. 

Pinta dos gráficas en dos cajas distintas, a la izquierda una recta con pendiente positiva de 3 que pase por (0,0) y a la derecha una recta con pendiente negativa de 3 que pase por (0,-5). 

Elige la precisión en el eje X que desees.  

```python
x = np.linspace(0, 5, 100)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# primera recta
ax1.plot(x, 3 * x, color='blue')
ax1.grid(True)

# segunda recta
ax2.plot(x, -3 * x , color='red')
ax2.grid(True)

# Mostrar la gráfica
plt.tight_layout()
plt.show()
```

3. Fija el eje X entre 0 y 5 y el eje Y entre -15 y 15

```python
x = np.linspace(0, 5, 100)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# primera recta
ax1.plot(x, 3 * x, color='blue')
ax1.set_xlim(0, 5)
ax1.set_ylim(-15, 15)
ax1.grid(True)

# segunda recta
ax2.plot(x, -3 * x , color='red')
ax2.set_xlim(0, 5)
ax2.set_ylim(-15, 15)
ax2.grid(True)

# Mostrar la gráfica
plt.tight_layout()
plt.show()
```

4. Llama al eje X "eje X", al eje Y "eje Y" y pon de títulos "recta sube" y "recta baja". Muestra dos etiquetas de ejeX pero solo una de eje Y.

```python
x = np.linspace(0, 5, 100)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# primera recta
ax1.plot(x, 3 * x, label='y = 3x', color='blue')
ax1.set_xlim(0, 5)
ax1.set_ylim(-15, 15)
ax1.set_title('Recta sube')
ax1.set_xlabel('eje X')
ax1.set_ylabel('eje Y')
ax1.grid(True)
ax1.legend()

# segunda recta
ax2.plot(x, -3 * x , label='y = -3x', color='red')
ax2.set_xlim(0, 5)
ax2.set_ylim(-15, 15)
ax2.set_title('Recta baja')
ax2.set_xlabel('eje X')
ax2.grid(True)
ax2.legend()

# Mostrar la gráfica
plt.tight_layout()
plt.show()
```

5. Usando una figura Y EJES POR SEPARADO, usa numpy para los valores del eje X entre 0 y 5. Pinta dos gráficas en dos cajas distintas, a la izquierda una recta con pendiente positiva de 3 que pase por (0,0) y a la derecha una recta con pendiente negativa de 3 que pase por (0,-5). Elige la precisión en el eje X que desees.  

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear la figura y los ejes (subgráficas)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Valores de X
x = np.linspace(0, 5, 100)

# Graficar en el primer eje
ax1.plot(x, 3 * x, label='y = 3x', color='blue')
ax1.set_title('Recta sube')
ax1.set_xlabel('eje X')
ax1.set_ylabel('eje Y')
ax1.grid(True)
ax1.legend()

# Graficar en el segundo eje
ax2.plot(x, -3 * x - 5, label='y = -3x - 5', color='red')
ax2.set_title('Recta baja')
ax2.set_xlabel('eje X')
ax2.grid(True)
ax2.legend()

# Establecer límites para los ejes
ax1.set_xlim(0, 5)
ax1.set_ylim(-15, 15)

ax2.set_xlim(0, 5)
ax2.set_ylim(-15, 15)

# Mostrar la gráfica
plt.tight_layout()  # Ajusta el espaciado entre subgráficas
plt.show()
```

19. Vamos a pintar un histograma de unas alturas generadas aleatoriamente. Crea 300 alturas entre 150 y 190 cm.

```python
alturas_rand = np.random.randint(150, 191, 300)

plt.hist(alturas_rand,bins=16,edgecolor='black')

plt.xlabel('Altura en cm')  
plt.ylabel('Numero de seres')
plt.title('Alturas random')  

plt.show()
```

```python
alturas_rand = np.random.randint(150, 191, 300)

hist, bins, patches = plt.hist(alturas_rand, bins=20, color='blue', edgecolor='black')

for patch, freq in zip(patches, hist):
    patch.set_facecolor(plt.cm.viridis(freq / max(hist)))

plt.xlabel('Altura en cm') 
plt.ylabel('Numero de seres')
plt.title('Alturas random')
```

```python
media = 170 
desviacion_estandar = 4
tamano = 300  

alturas_rand = np.random.normal(media,desviacion_estandar,tamano)

alturas_rand = np.clip(alturas_rand.round(), 150, 190)

hist, bins, patches = plt.hist(alturas_rand, bins=20, color='blue', edgecolor='black')

for patch, freq in zip(patches, hist):
    patch.set_facecolor(plt.cm.viridis(freq / max(hist)))

plt.xlabel('Altura en cm') 
plt.ylabel('Numero de seres')
plt.title('Alturas random')
```
